# Skills
[Back to Contents](#Contents)

```python
df.info()
```

```python
# Execute this first
def clean_skills():
    for i in range(len(df["Skills"])):
        if type(df["Skills"][i]) != str:
            continue
        df["Skills"][i] = df["Skills"][i].split(': ')[1:][0]
clean_skills()
```

```python
# Execute this second
def clean_skills1():
    for i in range(len(df['Skills'])):
        if type(df['Skills'][i]) != str:
            df['Skills'][i] = [df['Skills'][i]]
            continue
        df['Skills'][i] = df['Skills'][i].split(',')
clean_skills1()
```
